```python
import sys
import copy
import functools

import torch
import torch.nn as nnexit
import torch.nn.functional as F

import numpy as np
import matplotlib.pyplot as plt
import random
import math

from dataset import binary_counter_1d
```

```python
# clam mode --- requires director

pattern_size = 3
director = torch.randn( pattern_size)
director /= director.norm()
plt.imshow(director.reshape(1,-1))
plt.show()

set_size = 10
x, y = binary_counter_1d( set_size, 30, 3, 5, 'clam', director=director, exclusion=True, centered=False)
for mu in range(set_size):
    plt.imshow(x[mu])
    plt.show()
    print(x[mu].norm(), y[mu])
```

```python
# pair mode --- requires patterns

pattern_size = 3
p1 = torch.randn( pattern_size)
p1 /= p1.norm()

p2 = torch.randn( pattern_size)
p2 -= p1 * ( torch.dot( p1, p2).item())
p2 /= p2.norm()

angle = (45. / 180.) * math.pi
p2 = p1 * math.cos( angle) + p2 * math.sin( angle)

plt.imshow(p1.reshape(1,-1))
plt.show()

plt.imshow(p2.reshape(1,-1))
plt.show()

patterns = torch.stack( [p1, p2])

set_size = 10
x, y = binary_counter_1d( set_size, 30, 3, 5, 'pair', patterns=patterns, exclusion=True, centered=False)
for mu in range(set_size):
    plt.imshow(x[mu])
    plt.show()
    print(x[mu].norm(), y[mu])
```

```python
# noisy mode --- requires director and noise

pattern_size = 3
director = torch.randn( pattern_size)
director /= director.norm()
plt.imshow(director.reshape(1,-1))
plt.show()

plt.imshow(-director.reshape(1,-1))
plt.show()

noise = 0.1

set_size = 10
x, y = binary_counter_1d( set_size, 30, 3, 5, 'noisy', director=director, noise=noise, exclusion=True, centered=False)
for mu in range(set_size):
    plt.imshow(x[mu])
    plt.show()
    print(x[mu].norm(), y[mu])
```
